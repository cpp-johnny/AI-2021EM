```python
# game
from random import random
import math

# Function to Print Table
def printtable(L):
    print('#'*61)
    print('#####  01  #  02  #  03  #  04  #  05  #  06  #  07  #  08  #')
    print('#'*61)
    for i in range(8):
        print('#'+ ' '*(2-len(str(i*8))) + str(i*8)+" #", end = "")
        for j in range(8):
            print(' ' + str(L[8*i+j]) + ' '*(5-len(str(L[8*i+j])))+'#', end="")
        print('')
        print('#'*61)

# Initialize Starting Table
# L is an array of numbers representing the numbers involved
# primes are the set of numbers to be multiplied to generate the starting tableau
L = [1] * 64
primes = [2,3,5,7]
p = len(primes)

# Setup Starting Table, there maybe the unlucky time where it loops forever
while L.count(1) != 0:
    if L.count(1) > 32:
        i = int(random()*64)
        j = int(random()*64)
    else:
        i = L.index(1)
        if L.count(1) == 1:
            j = 0
        else:
            j = L[i+1:].index(1) + i+1

    k = primes[int(random()*p)]
    if i != j:
        if L[i] > 50 or L[j] > 50:
            pass
        else:
            L[i] *= k
            L[j] *= k

printtable(L)

while L.count(1) < 63:
    #a = int(input("x-ccord for Selection 1? "))
    #b = int(input("y-ccord for Selection 1? "))
    #c = int(input("x-ccord for Selection 2? "))
    #d = int(input("y-ccord for Selection 2? "))
    #if 8*a + b - 8*c - d == 0:
    #    print('Same Cell Selected')
    #else:
    #    number = math.gcd(L[8*(a-1)+b-1],L[8*(c-1)+d-1])
    #    L[8*(a-1)+b-1] = int(L[8*(a-1)+b-1]/number)
    #    L[8*(c-1)+d-1] = int(L[8*(c-1)+d-1]/number)

    a = input("1st Cell? (1-64) ")
    try:
        a=int(a)
    except ValueError:
        a = 1
    b = input("2nd Cell? (1-64) ")
    try:
        b = int(b)
    except ValueError:
        b = 1

    if a == b:
        print('Same Cell Selected')
    else:
        number = math.gcd(L[a-1],L[b-1])
        L[a-1] = int(L[a-1]/number)
        L[b-1] = int(L[b-1]/number)

    printtable(L)

if L.count(1) == 64:
    print('Win')
else:
    print('Lose')

```
